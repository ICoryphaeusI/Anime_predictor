```python
data = data[data['Rate'] != 'Не смог']
data['Rate'] = data['Rate'].astype(float)
data = data[~data['Episodes'].str.contains("/", na=False)]
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
dicts_type = {}

label.fit(data.Type.drop_duplicates())
dicts_type['Type'] = list(label.classes_)
data['Type_encoder'] = label.transform(data['Type'])

# Ниже приведен ручной вариант выполнения.
# Столбец с возрастным ограничением был обработан вручную из-за принудительного уменьшения категориальных признаков

# type_mapping = {'TV Сериал':0, 'Фильм':1, 'Спешл':2, 'ONA':3, 'OVA':4, 'Тип не указан':5}
# data['Type_map'] = data['Type'].map(type_mapping)
# print(len(type_mapping))

dicts_episodes = {}
label.fit(data.Episodes.drop_duplicates())
dicts_episodes['Episodes'] = list(label.classes_)
data['Episodes_encoder'] = label.transform(data['Episodes'])

dicts_length = {}
label.fit(data['Episode length'].drop_duplicates())
dicts_length['Episode length'] = list(label.classes_)
data['Length_encoder'] = label.transform(data['Episode length'])

age_limit_mapping = {'PG-13':0, 'G':1, 'R-17':2, 'R+':3, 'PG':4, 'Не указано':5, 'R-17. В РФ только по достижению 18 лет.':2, 'PG-13. В РФ только по достижению 18 лет.':0, 'R+. В РФ только по достижению 18 лет.':3, 'PG. В РФ только по достижению 18 лет.':4}
data['Age_limit_map'] = data['Age Limit'].map(age_limit_mapping)

dicts_studio = {}
label.fit(data.Studio.drop_duplicates())
dicts_studio['Studio'] = list(label.classes_)
data['Studio_encoder'] = label.transform(data['Studio'])

dicts_author = {}
label.fit(data.Author.drop_duplicates())
dicts_author['Author'] = list(label.classes_)
data['Author_encoder'] = label.transform(data['Author'])

# Список со всеми жанрами
genres_list = ['Сёнен', 'Сёнен-ай', 'Сэйнэн', 'Сёдзё', 'Сёдзё-ай', 'Дзёсей', 'Комедия', 'Романтика', 'Школа', 'Безумие', 'Боевые искусства', 'Вампиры', 'Военное', 'Гарем', 'Гурман', 'Демоны', 'Детектив', 'Детское', 'Драма', 'Игры', 'Исторический',
               'Космос', 'Магия', 'Машины', 'Меха', 'Музыка', 'Пародия', 'Повседневность', 'Полиция', 'Приключения', 'Психологическое', 'Работа', 'Самураи', 'Сверхъестественное', 'Спорт', 'Супер сила', 'Ужасы', 'Фантастика', 'Фэнтези', 'Экшен',
               'Этти', 'Триллер', 'Эротика', 'Хентай', 'Яой', 'Юри']

for genre in genres_list:
    data[genre] = data['Genres'].apply(lambda x: 1 if genre in x else 0)
```
